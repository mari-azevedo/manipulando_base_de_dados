# Manipulando base de dados

```python
import gdown
import pandas as pd

url = 'https://drive.google.com/uc?id=1R1nvX6MocByvYBvMs5ujUhbdpW0gRe_X'
output = 'dados_credito.csv'
gdown.download(url, output, quiet=False)

credito = pd.read_csv('/content/dados_credito.csv', sep=';', decimal=',')
credito.head(10)
```

Verificando e corrigindo os nomes das colunas

```python
print('Antes:',credito.columns)
credito.columns = ['id', 'limites', 'sexo', 'escolaridade', 'status_civil', 'idade', 'pagamento', 'valor_divida', 'valor_pago']
print('Depois:',credito.columns)
```

Verificando e removando valores faltantes no banco

```python
print( 'Antes:',credito.shape)
print( 'Linhas com valores faltantes:',credito[credito.isna().any(axis=1)].shape)
display(credito[credito.isna().any(axis=1)])
credito = credito.dropna(how='any')
print( 'Depois de remover:',credito.shape)
```

As variáveis limites, valor_divida e valor_pago estão em dólar taiwanês.
Assumi que cada dólar taiwânes valha R$0,20. Transformei essas variáveis para que o valor esteja em reais (moeda)

```python
print("DataFrame original:")
print(credito[['limites', 'valor_divida','valor_pago']].head(5))
credito.loc[:,['limites','valor_divida','valor_pago']] *= 0.2
print("\nDataFrame atualizado:")
print(credito[['limites', 'valor_divida','valor_pago']].head(5))
```

O banco estava codificado. Criei variáveis que traduziram os códigos. No banco de dados "credito", criei as seguintes variáveis:

###sexo2:
Se sexo = 1, então sexo2 = Masculino
Se sexo = 2, então sexo2 = Feminino

###escolaridade2:
Se escolaridade <= 2, então escolaridade2 = Ensino superior ou maior
Se escolaridade >= 3, então escolaridade2 = Ensino médio ou menor

### status_civil2:
Se status_civil = 1, então status_civil2 = Casado(a)
Se status_civil = 2, então status_civil2 = Solteiro(a)
Se status_civil = 3, então status_civil2 = Outros

### pagamento2:
Se pagamento < 0, então pagamento2 = Antecipado
Se pagamento = 0, então pagamento2 = Em dia
Se pagamento > 0, então pagamento2 = Atrasado

```python
import numpy as np

# Definindo as condições e os valores
condições_sexo2 = [(credito['sexo']== 1),(credito['sexo']== 2)]
valores_sexo2 = ["Masculino","Feminino"]

condições_escolaridade2 = [(credito['escolaridade']<=2),(credito['escolaridade']>=3)]
valores_escolaridade2 = ["Ensino superior ou maior", "Ensino medio ou menor"]

condições_civil2 = [(credito['status_civil']==1),(credito['status_civil']==2),(credito['status_civil']==3)]
valores_civil2 = ["Casado(a)","Solteiro(a)","Outros"]

condições_pagamento2 = [(credito['pagamento']<0),(credito['pagamento']==0),(credito['pagamento']>0)]
valores_pagamento2 = ["Antecipado","Em dia","Atrasado"]

# Adicionando as colunas
credito['sexo2'] = np.select(condições_sexo2, valores_sexo2)
credito['escolaridade2'] = np.select(condições_escolaridade2, valores_escolaridade2)
credito['status_civil2'] = np.select(condições_civil2, valores_civil2)
credito['pagamento2'] = np.select(condições_pagamento2, valores_pagamento2)

# Imprimindo
credito.head(10)
```

Organizei a base de dados na seguinte ordem: id, limites, sexo2, escolaridade2, status_civil2, idade, pagamento2, valor_divida, valor_pago. Descartei as variáveis sexo, escolaridade, status_civil, e pagamento.

```python
credito = credito[['id', 'limites', 'sexo2', 'escolaridade2', 'status_civil2', 'idade', 'pagamento2', 'valor_divida', 'valor_pago']]
credito.head(10)
```

Crie a variável debito = valor_divida - valor_pago.

```python
credito['debito'] = credito['valor_divida'] - credito['valor_pago']
credito.head(10)
```

verifiquei quantos clientes possuem débito negativo (ou seja, crédito).

```python
# 8 com o débito negativo
print(credito[credito['debito']<0].shape)
display(credito[credito['debito']<0])
```

Deixando apenas clientes com débito

```python
print('Antes: ',credito.shape)
credito = credito[credito['debito']>=0]
print('Depois: ',credito.shape)
credito.head(10)
```
